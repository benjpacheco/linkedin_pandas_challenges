# Grouping data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.info()
```

## Average amount of each product bought by each group of customers

```python
amt_bought = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]
```

```python
df.groupby(['Marital_Status']).mean()[amt_bought]
```

## Grouping by multiple columns

```python
df.groupby(['Education', 'Marital_Status'])['Income'].agg(['median', 'mean'])
```

## Applying a custom aggregate function

```python
def topFive(df, n=5, column='NumWebPurchases'):
    return df.sort_values(by=column, ascending=False).head(5)
```

```python
df.groupby('Education')['NumWebPurchases', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases'].apply(topFive)
```
